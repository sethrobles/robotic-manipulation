```
# Compute the list of files that will be added and removed.
from glob import glob
from os.path import basename

def archive_path(base):
    return f"/opt/drake/share/drake/tutorials/{base}"

def work_path(base):
    return f"/work/{base}"

# Absolute paths.
released_paths = sorted(glob(archive_path("*.ipynb")))

# File names only, no paths.
released = [basename(x) for x in released_paths]

def print_list(heading, items):
    print(heading)
    if not items:
        print(" (None)")
    for item in items:
        print(f"- {item}")
    print()

print_list("Released tutorials:", released)
```

**Inspect the above list of "Released tutorials" before proceeding! There should be approximately two dozen files listed (the set of tutorials). These are the tutorials that you'll be deploying.**

```
import shutil
yyyymmdd = open('/opt/drake/share/doc/drake/VERSION.TXT').read().split()[0]
print(f"You are about to deploy version {yyyymmdd}.")
print(f"Type 'yes' in the little box below to copy the notebooks, or anything else to cancel.")
print("Note that the little box will fight you for focus.")
print("Keep clicking and you'll eventually win!")
response = input("Proceed")
print(f"The response was {repr(response)}.")
if response != "yes":
    raise RuntimeError("Cancelled")

for base in released:
    print(f"Copying {base}")
    shutil.copyfile(archive_path(base), work_path(base))
```
